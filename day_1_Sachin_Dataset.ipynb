# Q3. A match is chosen at a random. What is the probability that Sachin has scored a century in that match and India have won that match?

Using value_counts() function

```python
df[['century','Won']].value_counts()
```

Using crosstab() function

```python
# This table is called contingency table
pd.crosstab(index=df['century'],columns=df['Won'],margins=True)
```

Finding probability of winning and done century

```python
# Won -> True,and century -> True
#p=count of favourable outcomes/count of ss
(30/360)*100
```

```python
df[['century','Won']].value_counts()/len(df)
```

```python
sum((df['century']==True) & (df['Won']==True))/len(df)
```

```python
(sum((df['century']==True) & (df['Won']==True))/len(df))*100
```

There is 8% chance that Sachin has scored a century and India have won that match
